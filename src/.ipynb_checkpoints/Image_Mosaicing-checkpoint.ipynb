# Question 1

### Functions defined:
#### Stitch - used to stitch any two images
#### Ft_matches - feature matching
#### RandCons (RANSAC - Random sample consensus) - Ideal Homography matrix calculated after 1000 iterations
#### Homography - Returns the homography matrix
#### Output - Shows the stitched images

```python
def stitch(*img):
#    img1 = img[0]
#    img2 = img[1]
    points1 = ft_matches(*img)
    ran = RandCons(points1)
#     print(ran)
    panorama_size = (2 * max(img[1].shape[1],img[0].shape[1]), 2 * max(img[0].shape[0], img[1].shape[0]))
    out = cv2.warpPerspective(img[1], ran, panorama_size)
    out[0:img[0].shape[0], 0:img[0].shape[1], :] = img[0]
    return img_crop(out)



def ft_matches(*img):
    points = np.array([])
    sift = cv2.ORB_create()
    (key_point_1, desc_1), (key_points_2, desc_2) = sift.detectAndCompute(img[0],None), sift.detectAndCompute(img[1],None)

    #feature matching
    bf_match = cv2.BFMatcher()
    matches = bf_match.knnMatch(desc_1, desc_2, k = 2)

    for match1,match2 in matches:
        if match1.distance < 0.75*match2.distance:
            (x1,y1), (x2,y2) = key_point_1[match1.queryIdx].pt, key_points_2[match1.trainIdx].pt
            if points.size > 0:
                points = np.vstack([points,[x2,y2,x1,y1]])
            elif points.size == 0:
                points = np.array([[x2,y2,x1,y1]])
                
    return points

def RandCons(points1,thresh = 0.85,dis = 5):
    maxInliers = -1
    homo_best = None
    point_start = 0
    min_error = 1e9
#    tot = points1.shape[0]
    req = 0.85 * points1.shape[0]
    for i in range(1000): #NUM_ITERS
        total_error = 0
#        err = 0
        ind = np.random.randint(points1.shape[0], size=4)
        random_points = points1[random.randrange(0, points1.shape[0]), :].reshape(1, -1)
        
        for i in range (3):
            rand = points1[random.randrange(0, points1.shape[0]), :].reshape(1, -1)
            random_points = np.concatenate((random_points, rand), axis=0)
        
        homo_mat = homography(random_points)
        ##calculate inliers
        ##Returns the total Inliers from the correspondaces using the given h matrix and which are below the threshold
        for i in range(points1.shape[0]):
            point1, point2 = (np.array([points1[i][point_start], points1[i][point_start+1], 1]).reshape(1, -1).T), (np.array([points1[i][point_start+2], points1[i][point_start+3], 1]).reshape(1, -1).T)
            err = np.sqrt(np.linalg.norm(point2 - ((homo_mat@point1)/(homo_mat@point1)[2])))  #point2_estimate = homo_mat@point1
            total_error += err
        
        if total_error < min_error:
            homo_best = homo_mat
            
        min_error = min(total_error, min_error)
            
    return homo_best



def homography(relations):
    var_shape = relations.shape[0]
    matrix_homo = np.zeros((2 * var_shape, 9))
    for i in range(0, relations.shape[0]):

        point = 2*i
        adj_point = point+1
        #        point2 = np.matrix([relations[i][2], relations[i][3], 1])
        matrix_homo[point,0:3] = np.matrix([relations[i][0], relations[i][1], 1]) * -1 #-point1
        matrix_homo[adj_point, 3:6] = np.matrix([relations[i][0], relations[i][1], 1]) * -1 #-point1
        
        matrix_homo[point,6:9] = np.matrix([relations[i][0], relations[i][1], 1]) * np.matrix([relations[i][2], relations[i][3], 1]).item(0)
        matrix_homo[adj_point, 6:9] = np.matrix([relations[i][0], relations[i][1], 1]) * np.matrix([relations[i][2], relations[i][3], 1]).item(1)
        
        

    U, D, VT = np.linalg.svd(matrix_homo)

    h = VT[-1].reshape(3, 3)
    homography_ans = h / h[2,2]
    
    return (homography_ans)

  
def trim(im):
    diff = ImageChops.difference(im, Image.new(im.mode, im.size, im.getpixel((0,0))))
    diff = ImageChops.add(diff, diff, 2.0, -100)
    if diff.getbbox() == True:
        return im.crop(diff.getbbox())
    
    
def img_crop(img):
    s1 = 0
    s2 = 0
    c = 0
    for j in range(0, img.shape[1]):
        for i in range(img.shape[0]):
            if img[i][j][2] == 0:
                if img[i][j][1] == 0:
                    if img[i][j][0] == 0:
                        break
            c = c + 1
        if ( c >= s1 ):
            s1 = c
        c = 0
   
    for i in range(0, img.shape[0]):
        for j in range(img.shape[1]):
            if img[i][j][2] == 0:
                if img[i][j][1] == 0:
                    if img[i][j][0] == 0:
                        break
            c = c + 1
        if c >= s2:
            s2 = c
        c = 0
            
    img = img[0:s1, 0 : s2, :]
    return img 
    

def output(image_list, output): 
    index_number = 1;
    for n, image in enumerate(image_list):
        a = plt.figure().add_subplot(2, np.ceil(len(image_list)/float(2)), n + 1)
        plt.title("Fig"+str(index_number))
        plt.imshow(image)
        index_number = index_number + 1;
    
    plt.figure().set_size_inches(np.array(plt.figure().get_size_inches()) * len(image_list))
    plt.show()
    
    plt.figure(figsize = (15,15))
    plt.title("Stitched Image")
    plt.imshow(output)
    plt.show()
```
